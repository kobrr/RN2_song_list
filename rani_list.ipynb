```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os

pd.set_option('display.max_rows', 30)

def rani(chromedriver_path):
    # 5 and 6 are Sat. and Sun.
    if datetime.datetime.now().weekday() not in [5,6]:
        options = webdriver.ChromeOptions()
        options.add_argument('--headless')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        driver = webdriver.Chrome(options=options, executable_path=chromedriver_path)
        driver.implicitly_wait(10)

        today = str(datetime.datetime.now().date()).replace('-','')
        url = 'http://www.radionikkei.jp/ranimusic/onair/?date={}'.format(today)
        driver.get(url)
        html = driver.page_source.encode('utf-8')
        driver.quit()
        soup = BeautifulSoup(html, "html.parser")

        date = url[-8:-4] + '-' + url[-4:-2] + '-' + url[-2:]
        df = pd.DataFrame({'datetime': [date + ' ' +  i.text.replace('OA ','') for i in soup.find_all('p', class_='oa-time')],
        'title': [i.text for i in soup.find_all('p', class_='oa-title')],
        'artist': [i.text for i in soup.find_all('p', class_='oa-artist')]})
        
        return df

    else:
        pass

if __name__ == "__main__":
    chromedriver_path = '**/chromedriver'
    df = rani(chromedriver_path)
    display(df[:30])
```
